```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from collections import defaultdict
```

```python
files = glob.glob('../Solution/Hard*.csv')
files.sort()
files
```

```python
a = np.array(files)
b = [0,3,5]
files = list(a[b])
files = dict(enumerate(files))
files
```

```python
dfs = defaultdict(int)
for indx, f in files.items():
    dfs[indx] = pd.read_csv(f)
```

```python
dfs[0][:5]
```

## Iterations vs Time

```python
ax = plt.subplot(111)

plt.plot(dfs[0]['iter'],dfs[0]['time'], label = "Policy Iteration")
plt.plot(dfs[1]['iter'],dfs[1]['time'], label = "Q-Learning L0.1 E0.7")
plt.plot(dfs[2]['iter'],dfs[2]['time'], label = "Value Iteration")
    
#plt.ylim(0, 20)
plt.xlim(0, 100)
plt.ylabel("Time (in ms)")
plt.xlabel("Iterations")
plt.title('Hard Comparison: Iterations vs. Time')
ax.legend()
plt.show()
```

## Iterations vs Steps

```python
ax = plt.subplot(111)

plt.plot(dfs[0]['iter'],dfs[0]['steps'], label = "Policy Iteration")
plt.plot(dfs[1]['iter'],dfs[1]['steps'], label = "Q-Learning L0.1 E0.7")
plt.plot(dfs[2]['iter'],dfs[2]['steps'], label = "Value Iteration")

#plt.ylim(20, 100)
plt.xlim(0, 100)
plt.ylabel("Steps")
plt.xlabel("Iterations")
plt.title('Hard Comparison: Iterations vs. Steps')
ax.legend()
plt.show()
```

## Iterations vs Convergence

```python
ax = plt.subplot(111)

plt.plot(dfs[0]['iter'],dfs[0]['convergence'], label = "Policy Iteration")
plt.plot(dfs[1]['iter'],dfs[1]['convergence'], label = "Q-Learning L0.1 E0.7")
plt.plot(dfs[2]['iter'],dfs[2]['convergence'], label = "Value Iteration")

#plt.ylim(0, 10)
plt.xlim(0, 100)
plt.ylabel("Convergence")
plt.xlabel("Iterations")
plt.title('Hard Comparison: Iterations vs. Convergence')
ax.legend()
plt.show()
```

```python
ax = plt.subplot(111)

plt.plot(dfs[0]['iter'],dfs[0]['reward'], label = "Policy Iteration")
plt.plot(dfs[1]['iter'],dfs[1]['reward'], label = "Q-Learning L0.1 E0.7")
plt.plot(dfs[2]['iter'],dfs[2]['reward'], label = "Value Iteration")

plt.ylim(-400, 75)
plt.xlim(0, 100)
plt.ylabel("Reward")
plt.xlabel("Iterations")
plt.title('Hard Comparison: Iterations vs. Reward')
ax.legend()
plt.show()
```
