```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from collections import defaultdict
```

```python
files = glob.glob('../Solution/Easy Q-Learning*.csv')
files.sort()
#File name beginining
nb = 27
#File name ending
ne = -4
#Output file
#name = 'tsp.csv'
```

```python
files = dict(enumerate(files))
files
```

```python
dfs = defaultdict(int)
for indx, f in files.items():
    dfs[indx] = pd.read_csv(f)
```

## Iterations vs Time

```python
ax = plt.subplot(111)

for indx, d in dfs.items():
    ax.plot(d[['iter']],d[['time']], label = files[indx][nb:ne])

plt.ylim(0, 1)
plt.xlim(0, 600)
plt.ylabel("Time (in ms)")
plt.xlabel("Iterations")
plt.title('Q-Learning: Iterations vs. Time')
ax.legend()
plt.show()
```

## Iterations vs Steps

```python
ax = plt.subplot(111)

for indx, d in dfs.items():
    ax.plot(d[['iter']],d[['steps']], label = files[indx][nb:ne])

plt.ylim(8, 200)
plt.xlim(0, 600)
plt.ylabel("Steps")
plt.xlabel("Iterations")
plt.title('Q-Learning: Iterations vs. Steps')
ax.legend()
plt.show()
```

## Iterations vs Convergence

```python
ax = plt.subplot(111)

for indx, d in dfs.items():
    ax.plot(d[['iter']],d[['convergence']], label = files[indx][nb:ne])

plt.ylim(0, 10)
plt.xlim(0, 600)
plt.ylabel("Convergence")
plt.xlabel("Iterations")
plt.title('Q-Learning: Iterations vs. Convergence')
ax.legend()
plt.show()
```

## Iterations vs Reward

```python
ax = plt.subplot(111)

for indx, d in dfs.items():
    ax.plot(d[['iter']],d[['reward']], label = files[indx][nb:ne])

plt.ylim(-100, 100)
plt.ylabel("Reward")
plt.xlabel("Iterations")
plt.title('Q-Learning: Iterations vs. Reward')
ax.legend()
plt.show()
```
